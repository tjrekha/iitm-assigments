```python
from transformers import AutoTokenizer

sentence = "LLMs are powerful tools for natural language understanding."

# Load a pre-trained tokenizer (e.g., 'bert-base-uncased')
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Tokenize the sentence
encoded_input = tokenizer(sentence, return_tensors='pt', add_special_tokens=True)

# Get tokens
tokens = tokenizer.convert_ids_to_tokens(encoded_input['input_ids'][0])

# Get token IDs
token_ids = encoded_input['input_ids'][0].tolist()

# Get sequence length
sequence_length = len(token_ids)

print(f"Original sentence: {sentence}")
print(f"\nTokens: {tokens}")
print(f"\nToken IDs: {token_ids}")
print(f"\nSequence length: {sequence_length}")
```
