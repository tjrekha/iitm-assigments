## Load a Small LLM

### Subtask:
Load a small pre-trained language model, specifically 'distilgpt2', using the Hugging Face transformers library. This step will also load the corresponding tokenizer.


**Reasoning**:
To load the 'distilgpt2' model and its tokenizer, I need to import the necessary classes from the `transformers` library and then use their `from_pretrained` methods with the specified model name.



```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# Define the model name
model_name = 'distilgpt2'

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Load the model
model = AutoModelForCausalLM.from_pretrained(model_name)

print(f"Successfully loaded tokenizer and model for '{model_name}'.")
```

**Reasoning**:
The previous step successfully loaded the 'distilgpt2' model and tokenizer. The next logical step, according to the overall task, is to generate text continuations using the loaded model.



```python
prompt = 'AI is transforming industries by …'

# Encode the input prompt
inputs = tokenizer(prompt, return_tensors='pt')

# Generate text
generated_sequences = model.generate(
    inputs['input_ids'],
    max_length=50,  # Limit the length of generated text
    num_return_sequences=3, # Generate three continuations
    do_sample=True, # Enable sampling for more varied outputs
    top_k=50, # Consider top 50 tokens for sampling
    top_p=0.95, # Use nucleus sampling
    no_repeat_ngram_size=2, # Prevent repeating n-grams
    pad_token_id=tokenizer.eos_token_id # Set pad token for DistilGPT2
)

print(f"Generated text continuations for prompt: '{prompt}'\n")

# Decode and print the generated sequences
for i, seq in enumerate(generated_sequences):
    decoded_sequence = tokenizer.decode(seq, skip_special_tokens=True)
    print(f"Continuation {i+1}: {decoded_sequence}\n")
```
