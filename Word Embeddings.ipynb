```python
import gensim.downloader as api
word_vectors = api.load('glove-wiki-gigaword-50')
print("GloVe model 'glove-wiki-gigaword-50' loaded successfully.")
```

```python
target_words = ['king', 'queen', 'diamond']
print(f"Target words list created: {target_words}")

for word in target_words:
    if word in word_vectors:
        print(f"\n--- Analysis for '{word}' ---")
        # Display the first 10 values of its vector
        print(f"Vector (first 10 values): {word_vectors[word][:10]}")

        # Find and display the top 5 most similar words
        similar_words = word_vectors.most_similar(word, topn=5)
        print("Top 5 most similar words:")
        for similar_word, score in similar_words:
            print(f"- {similar_word}: {score:.4f}")
    else:
        print(f"\n--- Warning: '{word}' not found in vocabulary. ---")
```
